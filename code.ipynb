```python
df1=pd.DataFrame({'rooms':df['total_rooms'],'income':df['median_income'],'age':df['housing_median_age']})
df1
df2=pd.DataFrame({'price':df['median_house_value']})
df2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error
l=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df1,df2)
```

```python
x_train,y_train,x_test,y_test
```

```python
l.fit(x_train,y_train)
```

```python
l.coef_[:,0]
```

```python
y_pred=0
for i in range(3):
    y_pred=y_pred+l.coef_[:,i]*x_train.iloc[:,i]
y_pred+l.intercept_
```

```python
y_pred=l.predict(x_test)
y_pred
```

```python
l.score(x_test,y_test)*100
```

```python
mean_absolute_error(y_test,y_pred),np.sqrt(mean_absolute_error(y_test,y_pred))
```
